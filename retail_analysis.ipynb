# Retail Analysis with PostgreSQL

Exemplo de conexão ao banco e consultas analíticas.

```
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt

conn = psycopg2.connect(
    host='localhost',
    dbname='retail',
    user='postgres',
    password='postgres'
)

df = pd.read_sql("SELECT date_trunc('month', order_date) AS mes, SUM(total_amount) as vendas FROM orders WHERE order_status='completed' GROUP BY 1 ORDER BY 1", conn)
df.plot(x='mes', y='vendas', kind='bar', figsize=(10,5))
plt.title('Vendas Mensais')
plt.show()
```
